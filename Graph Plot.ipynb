```python
def scale(original_min, original_max, objective_min, objective_max, number_to_scale):
   original_scale = original_max - original_min
   original_percentage = ( (number_to_scale-original_min) * 100.0) / original_scale
   objective_scale = objective_max - objective_min
   objective_scale_from_zero = (objective_scale*original_percentage)/100.0
   return objective_min + objective_scale_from_zero
```

```python
def scale(original_min, original_max, objective_min, objective_max, number_to_scale):
   original_scale = original_max - original_min
   original_percentage = ( (number_to_scale-original_min) * 100.0) / original_scale
   objective_scale = objective_max - objective_min
   objective_scale_from_zero = (objective_scale*original_percentage)/100.0
   return objective_min + objective_scale_from_zero

import csv
from sets import Set
from igraph import *

x = [f for f in os.listdir('data/Co-word analysis result') if f.startswith('edgelist')]

for current_file in x:
    with open('data/Co-word analysis result/'+current_file, 'rb') as csvfile:
         reader = csv.DictReader(csvfile)
         nodes_list = list()
         edges_list = list()
         edges_labels_list = list()
         node_color_set = set()
         node_color_list = [None] * 500
         edges_color_list = list()         
         edges_width_list = list()
         min_width = float("inf")
         max_width = float("-inf")
         network1pass_amount_count = [0] * 100
         color_list = ["red", "darkgrey", "blue","whitesmoke", "pink", "yellow", "cyan", "magenta", "orange", "darkgreen", "darkblue", "darkred", "tomato"]
         networks_count = set()
         for row in reader:
            if(int(row['NetworkPass1'])>0 or int(row['NetworkPass2Node1'])>0):
                node_color_set.add(row['NetworkPass1'])
                networks_count.add(row['NetworkPass1'])
                if(row['Node1'] not in nodes_list):
                    nodes_list.append(row['Node1'])
                if(row['Node2'] not in nodes_list):
                    nodes_list.append(row['Node2'])
                edges_list.append((nodes_list.index(row['Node1']),nodes_list.index(row['Node2'])))
                edges_labels_list.append(row['Order'] + ", " + row['Co-occurrence'] + ", " + str(float(row['Strength'])))
                if(int(row['NetworkPass1'])>0 and int(row['NetworkPass2Node1']) is 0):               
                    node_color_list[nodes_list.index(row['Node1'])] = color_list[int(row['NetworkPass1'])]
                    node_color_list[nodes_list.index(row['Node2'])] = color_list[int(row['NetworkPass1'])]
                    network1pass_amount_count[int(row['NetworkPass1'])]+=1
                if(int(row['NetworkPass2Node1']) == 1 or int(row['NetworkPass2Node2']) == 1 ):
                    edges_color_list.append("green")
                else:
                    edges_color_list.append('grey')
                
                edges_width_list.append(int(row['Co-occurrence']))
                
                if(int(row['Co-occurrence']) > max_width):
                    max_width = int(row['Co-occurrence'])
                if(int(row['Co-occurrence']) < min_width):
                    min_width = int(row['Co-occurrence'])
         
         current_file_month = current_file[:-4].split("_")[1]

         min_size = float("inf")
         max_size = float("-inf")
         edges_size_list = [None] * 500
            
         ocurrence_dic = {}
            
         with open('data/Frequency of Occurrence/queryResults_SkillsOccurrence_'+current_file_month+".csv", 'rb') as csvfile_ocurrences:
            for line in csvfile_ocurrences :
                ocurrence = int(line.translate(None, '\"').replace(" ", "").split(",")[0])
                skill = line.translate(None, '\"').replace(" ", "").split(",")[1].rstrip()[38:]
                #if(skill in nodes_list):                
                if(ocurrence > max_size):
                    max_size = ocurrence
                if(ocurrence < min_size):
                    min_size = ocurrence                    
                ocurrence_dic[skill] = ocurrence
         
         for idx, val in enumerate(nodes_list):
             edges_size_list[idx]=scale(min_size,max_size,10,60,ocurrence_dic[val])
         
         bubbles_color_list = [0] * 100
         bubbles_position_list_x = [0] * 100
         bubbles_position_list_y = [0] * 100
         average_density=0
         average_centrality=0
         count = 0
         with open('data/Co-word analysis result/Network_'+current_file_month+".csv", 'rb') as csvfile_bubbles:
            for idx, line in enumerate(csvfile_bubbles) :
                if(idx > 0):
                    count+=1.0
                    line = line.split(',')
                    average_density = average_density + float(line[1])
                    average_centrality = average_centrality + float(line[2])
                    bubbles_color_list[int(line[0])]=float(line[0])
                    bubbles_position_list_x[int(line[0])]=float(line[2])
                    bubbles_position_list_y[int(line[0])]=float(line[1])
         average_centrality=average_centrality/count
         average_density=average_density/count
            
         #print average_centrality
         #print average_density
         #print count
        
    #print(nodes_list)
    #print(len(nodes_list))
    #print("Network pass:"+ str(node_color_set))
    #print(node_color_list)
    #print(edges_list)
    #print(edges_labels_list)

    nodes_list[nodes_list.index('ProjectManagement')]= "Project \n Managem."
    #nodes_list[nodes_list.index('ProjectManagement')]= "Project \n Managem."
    #nodes_list[nodes_list.index('ProjectManagement')]= "Project \n Managem."

    #edges_labels_list[1] = "1, 2, 0.011111111"

    g = Graph()
    g.add_vertices(len(nodes_list))
    g.vs["label"] = nodes_list
    g.vs["label_dist"] = [1.5]
    g.vs["size"] = edges_size_list
    g.vs["shape"] = ["circle"]
    #g.vs["label_color"] = ["white"]

    g.vs["color"] = node_color_list
    g.add_edges(edges_list)
    #g.es["label"] = edges_labels_list
    #g.es["size"] = [10]
    g.es["color"] = edges_color_list
    g.es["width"] = [scale(min_width,max_width,1,10,i) for i in edges_width_list]

    #g.vs["x"] = [5, 3.5, 2, 4, 5, 2, 3, 5]
    #g.vs["y"] = [1,   2, 2, 5, 5, 5, 5, 3]

    #['SQL', 'ProjectManagement', 'Excel', 'Analyst', 'Reporting', 'Analysis', 'Database', 'Design']

    #if not os.path.exists("plots/"+current_file[:-4]):
    #    os.makedirs("plots/"+current_file[:-4])
    
    #plot(g, margin = 80, bbox = (600, 600))
    #plot(g, "plots/"+current_file[:-4]+"/elisa_graph_fixed.png", margin = 80, bbox = (700, 300))
    #plot(g, "plots/"+current_file[:-4]+"/elisa_graph_circular.png", layout = g.layout("circular"), margin = 80, bbox = (800, 800))
    #plot(g, "plots/"+current_file[:-4]+"/elisa_graph_kk.png", layout = g.layout("kk"), margin = 80, bbox = (800, 800))
    plot(g, "plots/"+current_file[:-4]+"_graph_fr.png", layout = g.layout("fr"), margin = 80, bbox = (800, 800))
    #plot(g, "plots/"+current_file[:-4]+"/elisa_graph_large.png", layout = g.layout("large"), margin = 80, bbox = (800, 800))
    #plot(g, "plots/"+current_file[:-4]+"/elisa_graph_rt_circular.png", layout = g.layout("rt_circular"), margin = 80, bbox = (800, 800))
    bubble_plot(bubbles_position_list_x, 
                bubbles_position_list_y, 
                bubbles_color_list, 
                color_list, 
                network1pass_amount_count, 
                current_file_month,
                average_centrality,
                average_density,
                count)
```
